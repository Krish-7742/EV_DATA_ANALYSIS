```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import plotly.graph_objects as go
from scipy import stats
```

```python
# Load the EV performance dataset
df = pd.read_csv('ElectricCarData_Clean.csv')
```

```python
# Exploratory Data Analysis (EDA)
print(df.info())
print(df.describe())
```

```python
# Handle missing data
df = df.dropna()
```

```python
# Visualization 1: Scatter plot of Acceleration vs Top Speed
plt.figure(figsize=(10, 6))
plt.scatter(df['AccelSec'], df['TopSpeed_KmH'])
plt.xlabel('Acceleration (sec)')
plt.ylabel('Top Speed (km/h)')
plt.title('Acceleration vs Top Speed')
plt.grid()
plt.show()
```

```python
# Visualization 2: Bar chart of Battery Efficiency by Brand
plt.figure(figsize=(10, 6))
sns.barplot(x='Brand', y='Efficiency_WhKm', data=df)
plt.xlabel('Brand')
plt.ylabel('Efficiency (Wh/km)')
plt.title('Battery Efficiency by Brand')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 3: Histogram of Driving Range
plt.figure(figsize=(10, 6))
plt.hist(df['Range_Km'], bins=10)
plt.xlabel('Driving Range (km)')
plt.ylabel('Frequency')
plt.title('Distribution of Driving Range')
plt.show()
```

```python
# Visualization 4: Scatter plot of Price vs Driving Range
plt.figure(figsize=(10, 6))
plt.scatter(df['PriceEuro'], df['Range_Km'])
plt.xlabel('Price (Euro)')
plt.ylabel('Driving Range (km)')
plt.title('Price vs Driving Range')
plt.grid()
plt.show()
```
